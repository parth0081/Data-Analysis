#### REGION-WISE TERRORIST ATTACK GRAPH

```python
sns.set(rc={'figure.figsize':(25,10)})
sns.countplot(x='region_txt',data=data)
plt.title('TERRORIST ATTACKS VS REGIONS')
plt.xlabel('REGIONS')
plt.ylabel('REGION-WISE TERRORIST ATTACKS')
```

```python
data.head()
```

```python
victim_country=data['country_txt'].value_counts().reset_index()
top10_victim_country=victim_country[:11]
```

```python
top10_victim_country.columns=['country','no of terrorist attack']
```

### REGION-WISE TERRORIST ATTACKS

```python
plt.figure(figsize=(15,10))
plt.title('COUNTRY-WISE TERRORIST ATTACKS')
sns.barplot(x='no of terrorist attack',y='country',data=top10_victim_country)
plt.xlabel('NUMBER OF TERRORIST ATTACKS')
plt.ylabel('COUNTRY')
```

```python
india=data[(data['country_txt']=='India')]
indian_cities=india['city'].value_counts().reset_index()
indian_cities.columns=['city','number of terrorist attacks']
```

```python
indian_cities.drop(1,axis=0,inplace=True)
```

```python
top10_victim_indian_cities=indian_cities[:10]
```

```python
x=top10_victim_indian_cities['city']
y=top10_victim_indian_cities['number of terrorist attacks']
plt.title('CITY-WISE TERRORIST ATTACKS',fontdict={'color':'blue','size':20})
plt.xlabel('CITY',fontdict={'size':16})
plt.ylabel('NUMBER OF TERRORIST ATTACKS',fontdict={'size':16,'font':'bold'})
plt.bar(x,y,color='YELLOW')
```

### CITY-WISE TERRORIST ATTACKS IN INDIA PLOT

```python
data.head(1)
```

```python
worldwide_city=data.city.value_counts().reset_index()
worldwide_city.drop(0,axis=0,inplace=True)
worldwide_city.columns=['city','number of terrorist attacks']
top10_victim_worldwide_cities=worldwide_city[:10]
top10_victim_worldwide_cities.shape
```

```python
fig=px.bar(top10_victim_worldwide_cities,x='city',y='number of terrorist attacks',title='CITY-WISE TERRORIST ATTACKS (WORLDWIDE)',labels={'city':'CITY','number of terrorist attacks':'NUMBER OF TERRORIST ATTACKS'})
fig.show()
```

```python
yearwise=data['iyear'].value_counts().reset_index()
yearwise.columns=['year','no of terrorist attacks']
```

```python
yearwise=yearwise.astype({'year':'object'})
```

```python
yearwise.info()
```

```python
sns.barplot(y='no of terrorist attacks',x='year',data=yearwise)
plt.title('YEAR-WISE TERRORIST ATTACKS (WORLDWIDE)',fontdict={'size':30})
plt.xlabel('YEAR',fontdict={'size':20})
plt.ylabel('NUMBER OF TERRORIST ATTACK',fontdict={'size':20})
```

```python
fig=px.bar(yearwise,x='no of terrorist attacks',y='year',title='YEAR-WISE TERRORIST ATTACKS (WORLDWIDE)',labels={'year':'YEAR','no of terrorist attacks':'NUMBER OF TERRORIST ATTACK'})
fig.show()
```

```python
india=data[(data['country_txt']=='India')]
yearwise_india=india['iyear'].value_counts().reset_index()
yearwise_india.columns=['year','no of terrorist attacks']
yearwise_india=yearwise_india.astype({'year':'object'})
```
